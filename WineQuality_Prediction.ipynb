### Dataset

#### Rows Information

* The data set is consists of 1599 rows.

#### Columns information

* The data set is consists of 12 columns namely: Fixed acidity, Volatile acidity, Citric acid, Residual sugar, Chlorides, Free sulfur dioxide, Total sulfur dioxide, Density, pH, Sulphates, Alcohol, Quality.


1. Fixed acidity: It represents the fixed acidity level in wine, which is the non-volatile acids. These acids do not evaporate readily.
2. Volatile acidity: It refers to the volatile acidity level in wine, which is the amount of acetic acid present. Higher levels of volatile acidity can result in an unpleasant vinegar-like taste.
3. Citric acid: It represents the citric acid content in wine, which can add freshness and flavor to the wine.
4. Residual sugar: It denotes the residual sugar content remaining after fermentation. It contributes to the wine's sweetness.
5. Chlorides: It refers to the chloride content in wine. Higher chloride levels can contribute to a salty taste.
6. Free sulfur dioxide: It measures the free form of sulfur dioxide, which acts as an antimicrobial and antioxidant agent in wine.
7. Total sulfur dioxide: It represents the total amount of sulfur dioxide present in the wine. It includes both the free and bound forms.
8. Density: It measures the density of the wine, which is related to the sugar and alcohol content.
9. pH: It represents the pH value, which indicates the acidity or alkalinity of the wine. Lower pH values indicate higher acidity.
10. Sulphates: It denotes the sulphate content in wine. Sulphates can contribute to the wine's aroma and act as an antimicrobial agent.
11. Alcohol: It represents the alcohol content in wine, typically measured as a percentage by volume. It contributes to the wine's body and flavor.
12. Quality (Target variable): It is the quality rating of the wine, typically ranging from 3 to 9.
The quality is often assessed by sensory evaluations.

```python
def outlier_detection(df, column):
    z_scores = np.abs(stats.zscore(df[column]))
    df = df[(z_scores < 3)]
    print(column,'has',len(np.where(z_scores>3)[0]),'outliers')
    return df
```

```python
for column in df.columns:
    df = outlier_detection(df, column)
```

##### Handling outliers of Volume's column of the DataFrame: df

* #### Any value greater than upper limit or lower than lower limit will be considered as outlier.

```python
# finding the upper and lower limit of columns

def find_outlier_limits(col_name):
    Q1,Q3= df[col_name].quantile([.25,.75]) # Q1 --> first quartile 25% of data. Q3 --> third quartile 75% of data
    IQR = Q3-Q1                   # Inter quartile range
    low = Q1-(1.5* IQR)              # 1.5 = 150%
    high = Q3+(1.5* IQR)
    return (high,low)

high_facid,low_facid = find_outlier_limits('fixed acidity')
print('Fixed acidity: ','upper limit: ',high_facid,' lower limit: ',low_facid)

high_Vacid,low_Vacid = find_outlier_limits('volatile acidity')
print('Volatile acidity: ','upper limit: ',high_Vacid,' lower limit: ',low_Vacid)

high_Cacid,low_Cacid = find_outlier_limits('citric acid')
print('Citric acidity: ','upper limit: ',high_Cacid,' lower limit: ',low_Cacid)


high_Rsugar,low_Rsugar = find_outlier_limits('residual sugar')
print('Residual Sugar: ','upper limit: ',high_Rsugar,' lower limit: ',low_Rsugar)

high_chlorides,low_chlorides = find_outlier_limits('chlorides')
print('Chlorides: ','upper limit: ',high_chlorides,' lower limit: ',low_chlorides)



high_FSdioxide,low_FSdioxide = find_outlier_limits('free sulfur dioxide')
print('Free sulfur dioxide: ','upper limit: ',high_FSdioxide,' lower limit: ', low_FSdioxide)


high_TSdioxide,low_TSdioxide = find_outlier_limits('total sulfur dioxide')
print('Total sulfur dioxide: ','upper limit: ',high_TSdioxide,' lower limit: ', low_TSdioxide)


high_density,low_density = find_outlier_limits('density')
print('Density: ','upper limit: ',high_density,' lower limit: ', low_density)

high_Ph,low_Ph = find_outlier_limits('pH')
print('PH: ','upper limit: ',high_Ph,' lower limit: ', low_Ph)


high_sulphates,low_sulphates = find_outlier_limits('sulphates')
print('Sulphates: ','upper limit: ',high_sulphates,' lower limit: ', low_sulphates)


high_alcohol,low_alcohol = find_outlier_limits('alcohol')
print('Alcohol: ','upper limit: ',high_alcohol,' lower limit: ', low_alcohol)


high_quality,low_quality = find_outlier_limits('quality')
print('Quality: ','upper limit: ',high_quality,' lower limit: ', low_quality)



# Replacing outlier values with respective upper and lower limits

df.loc[df['fixed acidity']> high_facid,'fixed acidity'] = high_facid
df.loc[df['fixed acidity']< low_facid,'fixed acidity'] = low_facid


df.loc[df['volatile acidity']> high_Vacid,'volatile acidity'] = high_Vacid
df.loc[df['volatile acidity']< low_Vacid,'volatile acidity'] = low_Vacid


df.loc[df['citric acid']> high_Cacid,'citric acid'] = high_Cacid
df.loc[df['citric acid']< low_Cacid,'citric acid'] = low_Cacid


df.loc[df['residual sugar']> high_Rsugar,'residual sugar'] = high_Rsugar
df.loc[df['residual sugar']< low_Rsugar,'residual sugar'] = low_Rsugar


df.loc[df['chlorides']> high_chlorides,'chlorides'] = high_chlorides
df.loc[df['chlorides']< low_chlorides,'chlorides'] = low_chlorides


df.loc[df['free sulfur dioxide']> high_FSdioxide,'free sulfur dioxide'] = high_FSdioxide
df.loc[df['free sulfur dioxide']< low_FSdioxide,'free sulfur dioxide'] = low_FSdioxide

df.loc[df['total sulfur dioxide']> high_TSdioxide,'total sulfur dioxide'] = high_TSdioxide
df.loc[df['total sulfur dioxide']< low_TSdioxide,'total sulfur dioxide'] = low_TSdioxide


df.loc[df['density']> high_density,'density'] = high_density
df.loc[df['density']< low_density,'density'] = low_density

df.loc[df['pH']> high_Ph,'pH'] = high_Ph
df.loc[df['pH']< low_Ph,'pH'] = low_Ph


df.loc[df['sulphates']> high_sulphates,'sulphates'] = high_sulphates
df.loc[df['sulphates']< low_sulphates,'sulphates'] = low_sulphates


df.loc[df['alcohol']> high_alcohol,'alcohol'] = high_alcohol
df.loc[df['alcohol']< low_alcohol,'alcohol'] = low_alcohol

df.loc[df['quality']> high_quality,'quality'] = high_quality
df.loc[df['quality']< low_quality,'quality'] = low_quality

```
